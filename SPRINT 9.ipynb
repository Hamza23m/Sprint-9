```python
### S9-1.1 Claculadora del indice de masa corporal.

def calculadora_imc(peso, altura):
    valor_imc = peso / altura**2 
    return valor_imc

def clasificador_de_resultado(valor_imc):
    if valor_imc < 18.5:
        return f"Tu índice de masa corporal es {valor_imc:.1f}, tienes un peso bajo."
    elif valor_imc <= 24.9:
        return f"Tu índice de masa corporal es {valor_imc:.1f}, tienes un peso normal."
    elif valor_imc <= 29.9:
        return f"Tu índice de masa corporal es {valor_imc:.1f}, tienes sobrepeso."
    else:
        return f"Tu índice de masa corporal es {valor_imc:.1f}, tienes obesidad."

def main():
    try:
        peso = float(input("¿Cuál es tu peso en kg? "))
        altura = float(input("¿Cuál es tu altura en metros? "))
    except ValueError:
        print("El peso y la altura tienen que ser numeros") 
        return
    try:
        if peso <= 0 or altura <= 0:
            raise ValueError("El peso y la altura deben ser mayores que 0")
        if altura > 2.5:
            raise ValueError("La altura debe ser menor que 2.5 metros")
        resultado = clasificador_de_resultado(calculadora_imc(peso,altura))
        print(resultado)
    except ValueError as error: 
        print(error)
    
main()
```

```python
### S9-1.2 Conversor de temperaturas.
import pandas as pd 
def celsius(temperatura): 
    c = temperatura
    f = round((temperatura * 9/5 + 32),3)
    k = round((temperatura + 273.15),2)
    return c,f,k
def fahrenheit(temperatura): 
    f = temperatura
    c = round(((temperatura - 32) * 5/9),2)
    k = round(((temperatura - 32) * 5/9 + 273.15),2)
    return f,c,k
def kelvin(temperatura): 
    k = temperatura
    f = round(((temperatura - 273.15) * 9/5 + 32),2)
    c = round((temperatura - 273.15),2)
    return k,f,c

def conversor_de_unidades(temperatura,unidades):
    dict_temperaturas = {}
    unidades = unidades.strip().upper() 
    if unidades == "C":
        tempe_entrada, tempe_f, tempe_k = celsius(temperatura)
        dict_temperaturas["Celsius"] = tempe_entrada
        dict_temperaturas["Fahrenheit"] = tempe_f
        dict_temperaturas["Kelvin"]= tempe_k
    elif unidades == "F":
        tempe_entrada, tempe_c, tempe_k = fahrenheit(temperatura)
        dict_temperaturas["Fahrenheit"] = tempe_entrada
        dict_temperaturas["Celsius"] = tempe_c
        dict_temperaturas["Kelvin"]= tempe_k
    elif unidades == "K":
        tempe_entrada, tempe_f, tempe_c = kelvin(temperatura)
        dict_temperaturas["Kelvin"] = tempe_entrada
        dict_temperaturas["Fahrenheit"] = tempe_f
        dict_temperaturas["Celsius"] = tempe_c
    return dict_temperaturas

def main():
    lista_final = []
    cantidad = int(input("¿Cuantas temperaturas quieres convertir?"))
    if cantidad <= 0:
            print("la cantidad de temperaturas tiene que ser mayor que 0")
            return
    unidades = str(input("dime la unidad en la que está, C, F, o K")).strip().upper()
    if unidades not in {"C","K","F"}:
        print("las unidades tienen que ser una de las letras que se indican")
        return
    for n in range(cantidad):
        try:         
            temperatura = float(input(f"dime la temperatura {n +1}"))              
        except ValueError:
            print("La temperatura tiene que ser un numero")
            return 
        if unidades == "K" and temperatura < 0:
            print("los grados Kelvin han de ser un numero mayor o igual que 0")
            return
        lista_final.append(conversor_de_unidades(temperatura,unidades))
    return lista_final

datos = main()  
if datos is None:
    pass 
else:
    data_temperaturas = pd.DataFrame(datos)
    data_temperaturas
data_temperaturas
```

```python
# S9-1.3 Contador de palabras 

def recibe_texto():
    return input("excribe una frase o texto")
def limpia_texto(texto):
    """
    Esta función recibe un texto y lo limpia 
    devolviendo una lista de palabra en minusculas
    y sin ningun tipo de signo
    """
    signos = (("?"," "),("¿"," "),("("," "),
              (")"," "),("."," "),(":"," "),
              ("!"," "),("¡"," "),(";"," "),
              (","," "))
    for viejo,nuevo in signos:
        texto = texto.replace(viejo,nuevo)
    texto = texto.lower()
    palabras = texto.split()
    return palabras 
def contador_de_palabras(palabras):
    """
    Esta funcion recibe un palabras 
    y cuenta las veces que se repite 
    cada palabra y devuele un diccionario 
    con la palabra como calve y 
    el nº que se repite como valor"""

    dict_palabras = {}
    for palabra in palabras:
        if palabra not in dict_palabras:
            dict_palabras[palabra] = 1 
        else:
            dict_palabras[palabra] +=1 
    return dict_palabras
def media_longitud(palabras):
    total = 0
    for palabra in palabras: 
        total += len(palabra)
    media = total/len(palabras)
    return media 
def main():
    texto = recibe_texto()
    texto_limpio = limpia_texto(texto)
    if texto_limpio == []:
        print("Tienes que introducir un texto")
        return
    contador = contador_de_palabras(texto_limpio)
    print(f"El contador {contador}")
    print(f"La media de longitud es: {media_longitud(texto_limpio):.2f}")

main()
```

```python
# S9-1.4 Diccionario inverso 

def crea_diccionario():
    diccionario = {}
    try :
        cantidad = int(input("Cuantas claves y valores quieres introducir"))
        if cantidad <= 0:
            print("La cantidad tiene que ser mayor que 0")
            return
    except ValueError:
        print("La cantidad tiene que ser un numero")
        return
    for x in range(cantidad):
        clave = input(f"dame la clave {x+1}")
        valor = input(f"dame el valor{x+1}")
        diccionario[clave] = valor
    return diccionario

def invertir_diccionario(diccionario):
    inverso = {}
    for clave,valor in diccionario.items():
        if valor not in inverso:
            inverso[valor] = [clave]
        else:
            inverso[valor].append(clave)
    for valor in inverso:
        if len(inverso[valor]) == 1:
            inverso[valor] = inverso[valor][0]
    return inverso
def hay_repetidos(dic_inverso):
    for valor,clave in dic_inverso.items():
        if type(clave) == list:
            print(f"valores repetidos {clave} pertenecen a la clave ['{valor}']")
def main():
    original = crea_diccionario()
    if original is None:
        return
    inverso = invertir_diccionario(original)
    print(f"El diccionario original {original}")
    print(f"El diccionario inverso {inverso}")
    hay_repetidos(inverso)
        
main() 
```

```python
# S9-2.2 Converion de tipos de datos
import ast
def recibe_lista():
    lista  = []
    entrada = input("Dame una lista de numero, tanto en letra como numeros")
    lista = ast.literal_eval(entrada)
    return lista 

def conversor_float(lista):
    lista_de_floats = []
    lista_no_floats = []
    for x in lista: 
        if isinstance( x, (tuple,list)):
                for i in x:
                        if isinstance(i,(tuple,list)):
                            floats, no_floats = conversor_float(i)
                            lista_de_floats.extend(floats)
                            lista_no_floats.extend(no_floats)
                        else:
                            try:
                                num = float(i)
                                lista_de_floats.append(num)
                            except ValueError:
                                 lista_no_floats.append(i)                                                              
        else:
            try:
                num = float(x)
                lista_de_floats.append(num)
            except ValueError:
                 lista_no_floats.append(x)
    return lista_de_floats, lista_no_floats

def main():
    return conversor_float(
                recibe_lista())

main()
```

```python
# S9-3.1 Generador de contraseñas 
import string as st
import numpy as np
import pyperclip as pyp

def longitud_de_caracteres():
    longitud =  int(input("Cunatos caracteres quieres que tenga la contraseña?"
                         " Tiene que ser minimo 8"))
    while longitud < 8 :
        print("la contraseña tiene que tener minimo 8 caracteres")
        longitud =  int(input("Cunatos caracteres quieres que tenga la contraseña?"
                         " Tiene que ser minimo 8"))
    return longitud

def tipo_de_caracteres():
    mayusculas = input("quieres que contanga myusculas, si o no ?").lower()
    if mayusculas == "si" or mayusculas == "sí":
        mayusculas = True
    else:
        mayusculas = False
    minusculas = input("quieres que contanga minúsculas, si o no ?").lower()
    if minusculas == "si" or minusculas == "sí":
        minusculas = True
    else:
        minusculas = False
    numeros = input("quieres que contanga números, si o no ?").lower()
    if numeros == "si" or numeros == "sí":
        numeros = True
    else:
        numeros = False
    signos = input("quieres que contanga signos, si o no ?").lower()
    if signos == "si" or signos == "sí":
        signos = True
    else:
        signos = False
    return mayusculas,minusculas,numeros,signos

def validador_de_caracteres():
    while True:
        mayusculas,minusculas,numeros,signos = tipo_de_caracteres()
        if mayusculas or minusculas or numeros or signos:
            break
        print("La contraseña tiene que tener mínimo un tipo de los caracteres")
    return mayusculas,minusculas,numeros,signos     

def generador_contraseñas():
    list_contraseña = []
    mayus = [] 
    mayus.extend(st.ascii_uppercase)
    minus = []
    minus.extend(st.ascii_lowercase)
    nums = []
    nums.extend(st.digits)
    sign = ["-","_","$","#","?","¿","*","€","@","&","!","¡","%"]
    longitud = longitud_de_caracteres()
    mayusculas,minusculas,numeros,signos = validador_de_caracteres()
    caracteres = [(mayusculas,mayus),(minusculas,minus),(numeros,nums),(signos,sign)]
    diponibles = []
    for x,i in caracteres:
        if x == True:
                list_contraseña.extend(np.random.choice
                              (i))
                diponibles.extend(i)          
    faltan = longitud - len(list_contraseña)
    for x in range(faltan):
        list_contraseña.extend(np.random.choice(diponibles))
    return list_contraseña

def main():
    contraseña = generador_contraseñas()
    np.random.shuffle(contraseña)
    contraseña = "".join(contraseña)
    pyp.copy(contraseña)
    print("Contraseña copiada en el portapapeles")
    print(f"La contraseña es: {contraseña}")

main() 
```

```python
#S9-3.2 Procesamiento simple de datos. 

import pandas as pd

df_historico_partidos = pd.read_csv(r"C:\Users\Usuario\Downloads\historic partits.txt",
                                    sep="\t",
                                    header=None,
                                    names=["Equipo local","Resultado","Equipo visitante"])
df_historico_partidos[["Goles local","Goles visitante"]] =  df_historico_partidos["Resultado"].str.split("-", expand=True)
df_historico_partidos["Goles local"] = df_historico_partidos["Goles local"].astype(int)
df_historico_partidos["Goles visitante"] = df_historico_partidos["Goles visitante"].astype(int)
equipos = pd.concat([df_historico_partidos["Equipo local"],df_historico_partidos["Equipo visitante"]]).unique()
df_resultado_equipos = pd.DataFrame(equipos,columns=["Equipo"])
df_resultado_equipos["Goles a favor"] = 0
df_resultado_equipos["Goles en contra"] = 0
df_resultado_equipos["Puntos"] = 0

goles_local = df_historico_partidos.groupby("Equipo local")["Goles local"].sum()
goles_visitante = df_historico_partidos.groupby("Equipo visitante")["Goles visitante"].sum()
goles_no_local = df_historico_partidos.groupby("Equipo local")["Goles visitante"].sum()
goles_no_visitante = df_historico_partidos.groupby("Equipo visitante")["Goles local"].sum()

df_resultado_equipos["Goles a favor"] = df_resultado_equipos["Equipo"].map(goles_local.add(goles_visitante, fill_value=0))
df_resultado_equipos["Goles en contra"] = df_resultado_equipos["Equipo"].map(goles_no_local.add(goles_no_visitante, fill_value=0))
def contador_de_puntos_local(fila):
    if fila["Goles local"]>fila["Goles visitante"]:
        return 3
    elif fila["Goles local"]<fila["Goles visitante"]:
        return 0
    else:
        return 1

def contador_de_puntos_visitante(fila):
    if fila["Goles visitante"]>fila["Goles local"]:
        return 3
    elif fila["Goles visitante"]<fila["Goles local"]:
        return 0
    else:
        return 1

df_historico_partidos["Puntos local"] = 0
df_historico_partidos["Puntos visitante"] = 0
df_historico_partidos["Puntos local"] = df_historico_partidos.apply(contador_de_puntos_local,axis=1)
df_historico_partidos["Puntos visitante"] = df_historico_partidos.apply(contador_de_puntos_visitante,axis=1)

puntos_local = df_historico_partidos.groupby("Equipo local")["Puntos local"].sum()
puntos_visitante = df_historico_partidos.groupby("Equipo visitante")["Puntos visitante"].sum()

df_resultado_equipos["Puntos"] = df_resultado_equipos["Equipo"].map(puntos_local.add(puntos_visitante, fill_value=0))
df_resultado_equipos
```
